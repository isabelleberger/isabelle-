```python
DURATION = 1.5
OUT_GIF = 'out.gif'
VALID_EXTENSIONS = ('png', 'jpg')
```

```python
def create_gif(filenames, duration=DURATION):
    images = []
    for filename in filenames:
        images.append(imageio.imread(filename))
    imageio.mimsave(OUT_GIF, images, duration=duration)
```

```python
from time import sleep
import itertools
import random
```

```python
colours = 'Red Green Amber'.split()
rotation = itertools.cycle(colours)

def rg_timer():
	return random.randint(5,9)

def light_rotation(rotation):
	for colour in rotation:
		if colour == 'Amber':
			print('Caution! The light is %s.\n' % colour)
			sleep(5)
		elif colour == 'Red':
			print('Stop! The light is %s.\n' % colour)
			sleep(rg_timer())
		else:
			print('Go! The light is %s.\n' % colour)
			sleep(rg_timer())
```
